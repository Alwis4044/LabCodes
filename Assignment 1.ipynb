## Read any image. Display the histogram, Equalized histogram, and image with equalized histogram.

```python
import matplotlib.pyplot as plt
import cv2
#from google.colab.patches import cv2_imshow
```

```python
image_path = 'bird_image.jpeg'
image = cv2.imread(image_path)
window_image = 'image'
cv2.imshow(window_image,image)
cv2.waitKey()
```

```python
# Load the image and convert it to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
window_image = 'gray'
cv2.imshow(window_image,gray)
cv2.waitKey()
```

```python
# Calculate the histogram
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
```

```python
# Plot the histogram
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("Number of Pixels")
plt.plot(hist)
plt.xlim([0, 256])
plt.show()
```

```python
# Equalize the histogram
eq_hist = cv2.equalizeHist(gray)
```

```python
# Calculate the histogram
hist2 = cv2.calcHist([eq_hist], [0], None, [256], [0, 256])
```

```python
# Plot the histogram
plt.figure()
plt.title("Equalized Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("Number of Pixels")
plt.plot(hist2)
plt.xlim([0, 256])
plt.show()

window_image = 'eq_hist'
cv2.imshow(window_image,eq_hist)
cv2.waitKey()

window_image = 'gray1'
cv2.imshow(window_image,gray)
cv2.waitKey()

window_image = 'eq_hist1'
cv2.imshow(window_image,eq_hist)
cv2.waitKey()
 
```
